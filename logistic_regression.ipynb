#Credit Card Fraud Detection Using Logistic Regression

The project entails the analysis of a credit card transaction dataset for fraud detection. The dataset is highly imbalanced, and we use logistic regression to classify the transactions into fraudulent or legitimate.

---
### **About the Dataset** (information published by the dataset collector)
The dataset consists of credit card transactions made by European cardholders in September 2013. The following are some facts about the dataset:
- 284,807 transactions over two days.
- Only 492 transactions (0.172%) are fraudulent.
The feature's type is numerical scaled with PCA except 'Time' and 'Amount'.
The Target Variable:
The 'Class':
  • 0 -> Valid transaction
  • 1 -> Fraudulent transaction

---
### **Goal**
The task will be to work out a logisitic regression based classifier of fraudulent cases. And afterwards perform the relevant evaluations.

```python
#importing Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

## **1. Loading and Exploring the Dataset**
Here, we load the dataset and explore its structure, including basic information, checking for missing values, and understanding class distribution.


```python
#import necessary libraries
import os
import pandas as pd

# check if the dataset is already downloaded
dataset_path = 'creditcard.csv'
if not os.path.exists(dataset_path):
    #download the dataset using Kaggle API
    os.system('kaggle datasets download -d mlg-ulb/creditcardfraud --unzip')

# load the dataset
credit_card_data = pd.read_csv(dataset_path)

# display the first 5 rows
credit_card_data.head()
```

```python
#checking for missing values
credit_card_data.isnull().sum()
```

```python
#class distribution (0 = Legitimate, 1 = Fraudulent)
credit_card_data['Class'].value_counts()
```

## **2. Data Analysis**
We separate the dataset into legitimate and fraudulent transactions and compare statistical measures for these two groups.


```python
#separating legitimate and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

# displaying their shapes
print("Legitimate transactions:", legit.shape)
print("Fraudulent transactions:", fraud.shape)
```

```python
# statistical summary of 'Amount' for legitimate transactions
legit.Amount.describe()
```

```python
#statistical summary of 'Amount' for fraudulent transactions
fraud.Amount.describe()
```

```python
#comparing mean values for both classes
credit_card_data.groupby('Class').mean()
```

## **3. Balancing the Dataset**
The dataset is highly imbalanced. To address this, we undersample the legitimate transactions to match the number of fraudulent ones.


```python
#randomly sample legitimate transactions
legit_sample = legit.sample(n=492)

# combine the balanced legitimate transactions with fraudulent ones
new_dataset = pd.concat([legit_sample, fraud], axis=0)

#verifying the new dataset
new_dataset['Class'].value_counts()
```

## **4. Splitting the Dataset**
The dataset is split into training and testing sets (80-20 split), maintaining class balance using stratified sampling.


```python
# splitting features (X) and target (Y)
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']

# splitting into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

#displaying shapes
print("Original data shape:", X.shape)
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

## **5. Model Training and Evaluation**
We train a logistic regression model and evaluate its performance on both the training and test datasets.


```python
#initializing the Logistic Regression model
model = LogisticRegression()

#training the model
model.fit(X_train, Y_train)

#making predictions on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

print("Accuracy on Training Data:", training_data_accuracy)
```

```python
#making predictions on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print("Accuracy on Test Data:", test_data_accuracy)
```

## **6. Results**
- **Training Accuracy**: 94.28%
- **Test Accuracy:** 92.89%

Despite the class imbalance, the logistic regression model still shows quite decent performance. Further actions could include more complex models or advanced techniques like precision-recall curves, feature engineering, and hyperparameter optimization, although the results are quite promising.
